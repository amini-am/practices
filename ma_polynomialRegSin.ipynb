```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
def generateData(begin, ends, count = 100):
    x = np.linspace(begin, ends, count)
    y = np.sin(x)
    return (x.reshape(len(x), 1), y)

def generateXPoly(x, n, escap_evens = True):    
    j = 1
    X = x.copy()
    for i in range(2, n+1):
        if escap_evens and i %2 == 0:
            continue
        X = np.c_[X, np.power(x, i)]
        j = j+1
    
    return X
```

```python
x, y = generateData(0, 2 * np.pi)
X = generateXPoly(x, 11)

xt, yt = generateData(-2 * np.pi, 0)
Xt = generateXPoly(xt, 11)

model = LinearRegression()
model.fit(X, y)

yp = model.predict(Xt)

print(metrics.mean_squared_error(yt, yp))

plt.plot(xt, yp)
```
